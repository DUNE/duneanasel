# RDF File Reading Example
This notebook demonstrates how to use the ROOT `RDataFrames` to read and analyze data stored in `CAF` files. It covers the following topics:

- Importing the necessary `duneanaobj` ROOT libraries and adding include paths and dynamic libraries
- Creating an instance of `RDataFrame` and loading a `CAF` file from a specified path
- Defining new variables using the `Define` method of `RDataFrame`. Note that the functions that define the new branches are defined in `RDF_functions.h` and declared using the `gInterpreter`
- Displaying histograms of muon (anti)neutrino energy in classic RDataFrame fashion
- Displaying histograms of primary particle momenta from muon neutrino interactions

### Importing modules, libraries and loading tree in RDataFrame

```python
import ROOT

# ROOT.gSystem.AddIncludePath("../../duneanasel/build/_deps/duneanaobj-src")
ROOT.gSystem.AddDynamicPath("../../build/_deps/duneanaobj-build/duneanaobj/StandardRecord")
ROOT.gSystem.Load("libduneanaobj_StandardRecord")

rdf = ROOT.RDataFrame("cafTree", "../../../work_folder/MicroProdN4p1_NDComplex_FHC.caf.full.light.spineonly.0002459.CAF.root")
```

### Printing column names

```python
columns = rdf.GetColumnNames()

for col in columns:
    print(f"{col}")
```

### Loading RDF utils

```python
ROOT.gInterpreter.Declare(""" #include "RDF_functions.h" """)
```

# Plotting neutrino energy 

### Defining new branches for plotting

```python
rdf = (rdf.Define("a_numu_E", "nu_E_by_pdg(rec.mc.nu, -14)")
        .Define("numu_E", "nu_E_by_pdg(rec.mc.nu, 14)"))
```

```python
rdf.Display({"numu_E", "a_numu_E"},2).Print()
```

### Simple plotting

```python
# PDG codes
ROOT.gStyle.SetOptStat(0)

# Create histograms of neutrino energy
hist_numu     = rdf.Histo1D(("numu_energy", "Neutrino Energy by Pdg;E_{#nu} [GeV];Events", 50, 0, 100), "numu_E")
hist_antinumu = rdf.Histo1D(("antinumu_energy", "Anti-Muon Neutrino Energy;E [GeV];Events", 50, 0, 100), "a_numu_E")

h_numu = hist_numu.GetValue()
h_antinumu = hist_antinumu.GetValue()

# Draw both on the same canvas
c = ROOT.TCanvas("c", "NuMu vs Anti-NuMu", 800, 600)
h_numu.SetLineColor(ROOT.kBlue)
h_numu.SetFillColor(ROOT.kBlue)
h_numu.SetFillStyle(3003)
h_antinumu.SetLineColor(ROOT.kRed)
h_antinumu.SetFillColor(ROOT.kRed)
h_antinumu.SetFillStyle(3004)


h_numu.Draw("HIST")
h_antinumu.Draw("HIST SAME")

# Legend
legend = ROOT.TLegend(0.6, 0.6, 0.8, 0.8)
legend.SetBorderSize(0)
legend.SetTextFont(122)
legend.SetTextSize(0.05)
legend.AddEntry(h_numu, "#nu_{#mu}", "f")
legend.AddEntry(h_antinumu, "#bar{#nu}_{#mu}", "f")
legend.Draw()

c.Draw()
```

# Primary particle momentum distribution

### Defining new branches for plotting

```python
rdf = (rdf.Define("numu_prim_pdg", "prim_pdg(rec.mc.nu, 14)")
          .Define("numu_prim_mu_p", "prim_p_by_pdg(rec.mc.nu, 14, 13)")
          .Define("numu_prim_neutron_p", "prim_p_by_pdg(rec.mc.nu, 14, 2112)")
          .Define("numu_prim_proton_p", "prim_p_by_pdg(rec.mc.nu, 14, 2212)")
          .Define("numu_prim_piplus_p", "prim_p_by_pdg(rec.mc.nu, 14, 211)")
          .Define("numu_prim_piminus_p", "prim_p_by_pdg(rec.mc.nu, 14, -211)")
          .Define("numu_prim_pi0_p", "prim_p_by_pdg(rec.mc.nu, 14, 111)"))
```

```python
rdf.Display({"numu_prim_pdg","numu_prim_mu_p","numu_prim_neutron_p","numu_prim_proton_p"},1).Print()
rdf.Display({"numu_prim_piplus_p","numu_prim_piminus_p","numu_prim_pi0_p"},1).Print()
```

### Simple plotting

```python
histo_list = ["numu_prim_proton_p","numu_prim_neutron_p","numu_prim_mu_p","numu_prim_piplus_p","numu_prim_piminus_p","numu_prim_pi0_p"]
clour_list = [ROOT.kBlue, ROOT.kRed, ROOT.kGreen, ROOT.kYellow, ROOT.kMagenta, ROOT.kCyan]
fill_style_list = [3001, 3002, 3003, 3004, 3005, 3006]
legend_list = ["Primary Proton", "Primary Neutron", "Primary Muon", "Primary Positive Pion", "Primary Negative Pion", "Primary Neutral Pion"]
hist_objects = []

c1 = ROOT.TCanvas("c1", "Primary Particle Momenta", 800, 600)
legend = ROOT.TLegend(0.5, 0.65, 0.8, 0.85)
legend.SetBorderSize(0)
legend.SetTextFont(42)
legend.SetTextSize(0.05)

for i, (histo, color, fill_style) in enumerate(zip(histo_list, clour_list, fill_style_list)):
    h = rdf.Histo1D((histo, "#nu_{#mu} Primary Particle Momentum;p [GeV/c];Events", 50, 0, 10), histo)
    hist = h.GetValue()  # this creates the actual TH1D
    hist.SetLineColor(color)
    hist.SetFillColor(color)
    hist.SetFillStyle(fill_style)

    hist_objects.append(hist)  # <-- keep it alive

    draw_option = "HIST" if i == 0 else "HIST SAME"
    hist.Draw(draw_option)
    legend.AddEntry(hist, legend_list[i], "f")

legend.Draw()
c1.Draw()
```
